```python
import shutil
import os
```

```python
def copy_and_rename(src_path, dest_path, original_name, new_name): 
    # Copy the file 
    new_path = rf"{dest_path}\{new_name}"
    copied_path = rf"{dest_path}\{original_name}"
    shutil.copy(src_path, dest_path)

    # Rename the copied file 
    os.rename(copied_path, new_path)
```

```python
src_path = r"D:\Github\python-coding-stuff\03_handling_files\01_resources\01_copy_file_and_change_name\01_original_folder\original_file.txt"
dest_path = r"D:\Github\python-coding-stuff\03_handling_files\01_resources\01_copy_file_and_change_name\02_new_folder"
new_name = "new_file.txt"
original_name = "original_file.txt"

copy_and_rename(src_path, dest_path, original_name, new_name)
```
