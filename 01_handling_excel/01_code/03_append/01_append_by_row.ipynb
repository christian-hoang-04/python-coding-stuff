```python
import openpyxl
```

```python
file_path = r"D:\Github\python-coding-stuff\01_handling_excel\00_dataset\03_append\test_02.xlsx"

wb = openpyxl.load_workbook(file_path)

sheet = wb.active

data = (
    (1, 2, 3), 
    (4, 5, 6)
)

for row in data: 
    sheet.append(row)

wb.save(file_path)
```
