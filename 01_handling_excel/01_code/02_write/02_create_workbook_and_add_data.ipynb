```python
# import openpyxl module 
import openpyxl 

wb = openpyxl.Workbook()

sheet = wb.active
```

```python
c1 = sheet.cell(row=1, column=1)

# writing values to cells 
c1.value = "Hello"
```

```python
c2 = sheet.cell(row=1, column=2)
c2.value = "World"
```

```python
c3 = sheet['A2']
c3.value = "Welcome"
```

```python
# B2 means column = 2 & row = 2. 
c4 = sheet['B2']
c4.value = "Everyone"
```
